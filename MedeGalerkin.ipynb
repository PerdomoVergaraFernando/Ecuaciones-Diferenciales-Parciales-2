```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Queremos resolver la ecuacion

$$\frac{\partial^2{u}}{\partial{x^2}} + \frac{\partial^2{u}}{\partial{y^2}} = c$$

```python
x,y,a,b,c,A = sp.symbols('x,y,a,b,c,A') #introduciomos una variable
u = sp.Function('u') #Esto es una funci√≥n
```

```python
u_1 = A*x*(a-x)*y*(b-y)
u_1
```

```python
E= u_1.diff(x,x) + u_1.diff(y,y) - c
E
```

```python
I = sp.integrate(E*u_1,(y,0,b), (x,0,a))
I
```

```python
I.simplify()
```

```python
alfa = sp.solve (I,A)
```

```python
alfa
```

```python
al=alfa[1]
al
```

```python
u_approx = u_1.subs(A,alfa[1])
u_approx
```
