#<font color=red> Esto es una grafica que muestra una Parabola y=x²


```python

import matplotlib.pyplot as plt
import numpy as np

# Generamos un rango de valores para x
x = np.linspace(-100, 100, 400)
# Calculamos los valores de y para cada x
y = x**2

# Creamos la gráfica
plt.plot(x, y, label="y = x^2", color="red")

# Añadimos título y etiquetas
plt.title("Gráfica de y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Agrgamos lineas horizontales y verticales en el origen
plt.axhline(0 ,color="black", linewidth=0.5, ls="-" )
plt.axvline(0, color="black", linewidth=0.5, ls="-" )

# Añadimos una leyenda
plt.legend()

# Mostramos la gráfica
plt.grid(True)
plt.show()
```
