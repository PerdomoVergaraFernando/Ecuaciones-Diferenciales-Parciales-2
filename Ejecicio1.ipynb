Esto es una grafica que muestra una Parabola

```python

import matplotlib.pyplot as plt
import numpy as np

# Generamos un rango de valores para x
x = np.linspace(-100, 100, 400)
# Calculamos los valores de y para cada x
y = x**2

# Creamos la gráfica
plt.plot(x, y, label="y = x^2")

# Añadimos título y etiquetas
plt.title("Gráfica de y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Añadimos una leyenda
plt.legend()

# Mostramos la gráfica
plt.grid(True)
plt.show()
```
