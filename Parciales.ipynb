```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necesario para 3D
from matplotlib import cm  # colormap

U=np.array([[0,2/3,2/3,0],
            [0,0.3611,0.5833,0],
            [0,0.1944,0.4166,8/9],
            [0,0,0,0]])

x = np.linspace(0, 2, 4)
y = np.linspace(0, 2, 4)
X, Y = np.meshgrid(x, y)

# Mapa de colores
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, U, levels=10, cmap=cm.viridis)
plt.colorbar(contour)
plt.title('Solucin numrica de un problema de Dirichlet')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, U, cmap=cm.viridis, edgecolor='k')
ax.set_title('Solucin de Laplace con Dirichlet')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
